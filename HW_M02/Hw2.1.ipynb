```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Зчитування даних
url = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8",
                       attrs={"class": "wikitable collapsible collapsed", "style": "text-align: right"},
                       match="Регіон"
                      )
table = url[0]
table
```

```python
# 1 Вивести перші рядки таблиці за допомогою методу head
table.head()
```

```python
# 2 Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
table.shape
```

```python
# 3 Заміна у таблиці значень "—" на значення NaN
table.replace('—', np.nan, inplace=True)
table
```

```python
# 4 Визначте типи всіх стовпців за допомогою dataframe.dtypes
table.dtypes
```

```python
# 5 Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
non_numeric_columns = ["2014", "2019"]
table[non_numeric_columns] = table[non_numeric_columns].apply(pd.to_numeric, errors='coerce')
table.dtypes
```

```python
# 6 Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_percentage = (table.isnull().sum() / len(table)) * 100
missing_percentage
```

```python
# 7 Видаліть з таблиці дані по всій країні, останній рядок таблиці
table = table.drop(table.index[-1])
table
```

```python
# 8 Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
values = {
    "1950": table["1950"].mean().round(1),
    "1960": table["1960"].mean().round(1),
    "1970": table["1970"].mean().round(1),
    "1990": table["1990"].mean().round(1),
    "2000": table["2000"].mean().round(1),
    "2012": table["2012"].mean().round(1),
    "2014": table["2014"].mean().round(1),
    "2019": table["2019"].mean().round(1),
}
table = table.fillna(value=values)
table
```

```python
# 9 Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
average_birth_rate_2019 = table["2019"].mean().round(1)
regions_higher_than_average_2019 = table[table["2019"] > average_birth_rate_2019][["Регіон", "2019"]]
regions_higher_than_average_2019
```

```python
# 10 У якому регіоні була найвища народжуваність у 2014 році?

region_highest_birth_rate_2014 = table.sort_values(["2014"], ascending=False)[["Регіон", "2014"]].head(1)
region_highest_birth_rate_2014
```

```python
# 11 Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
import seaborn as sns

# Set a Seaborn style
sns.set(style="whitegrid")

# P1 T11 (Enhanced Visualization)
plt.figure(figsize=(12, 8))
colors = sns.color_palette("pastel", n_colors=len(table))

# Plot a horizontal bar chart for birth rates in 2019
sns.barplot(x="2019", y="Регіон", data=table.sort_values("2019", ascending=False),
            palette=colors, dodge=False)

plt.xlabel("Народжуваність у 2019 році")
plt.ylabel("Регіон")
plt.title("Народжуваність по регіонах у 2019 році")
plt.show()
```

```python
# For this task  Visualization - 9 Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

plt.figure(figsize=(12, 6))
sns.barplot(x="2019", y="Регіон", data=regions_higher_than_average_2019, palette="muted")
plt.title('Регіони з народжуваністю вище середнього у 2019 році')
plt.xlabel('Народжуваність у 2019 році')
plt.ylabel('Регіон')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.stackplot(table.index, table["1950"], table["1960"], table["1970"], table["1990"], table["2000"], table["2012"], table["2014"], table["2019"],
              labels=["1950", "1960", "1970", "1990", "2000", "2012", "2014", "2019"], alpha=0.8)
plt.title('Внесок кожного року в народжуваність (1950-2019)')
plt.xlabel('Індекс регіону')
plt.ylabel('Народжуваність')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
